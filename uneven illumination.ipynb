```python
from skimage import filters, morphology
from matplotlib import pyplot as plt
```

```python
# remove noise with rank median (similar to histogram center)
selem = morphology.disk(5)
filtered = filters.rank.median(img, selem)
io.imshow(filtered, vmin=filtered.min(), vmax=filtered.max())
plt.colorbar();
```

```python
CS = plt.contour(filtered)
cbar = plt.colorbar(CS)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
```

```python
# interactive
#%matplotlib qt
# inline
%matplotlib inline

figure(figsize=(10,10))
x, y = np.arange(0, img.shape[1], 16), np.arange(0, img.shape[0], 16)
X, Y = np.meshgrid(x,y)
ax = plt.subplot(projection='3d')
ax.plot_surface(X, Y, img[::16,::16])
ax.set_xlabel('X')
ax.set_zlabel('Intensity')

ax.set_ylabel('Y')
ax.set_ylim3d(512,0)

ax.view_init(elev=10, azim=270);
```
