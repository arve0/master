```python
from skimage import io
img = io.imread('overview.png')
```

```python
%matplotlib inline
```

```python
io.imshow(img);
```

```python
from skimage import filters, morphology
```

```python
selem = morphology.disk(3)
img = filters.rank.median(img, selem)
```

```python
io.imshow(img[:1000, :1000])
```

```python
equalized = filters.rank.equalize(img, selem)
```

```python
filters.threshold_yen(equalized), filters.threshold_otsu(equalized)
```

```python
io.imshow(equalized[:2000,:2000] < 177)
```

```python
binary = equalized <= filters.threshold_otsu(equalized)
```

```python
selem = morphology.disk(2)
binary = morphology.binary_erosion(binary, selem)
```

```python
io.imshow(binary[:1000,:1000])
```

```python
selem = morphology.disk(4)
binary = morphology.binary_dilation(binary, selem)
```

```python
io.imshow(binary[:1000,:1000])
```

```python
selem = morphology.disk(4)
binary = morphology.binary_dilation(binary, selem)
io.imshow(binary[:1000,:1000])
```
