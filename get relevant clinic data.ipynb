```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_stata('data/klinisk_data.dta', convert_categoricals=False)
```

```python
for c in data.columns:
    print('-', c)
```

```python
new_df = pd.DataFrame()
```

```python
new_df['person_id'] = data.person_id
```

```python
new_df['ID_deltaker'] = data.ID_deltaker
```

```python
np.all(new_df == data.loc[:, ['person_id', 'ID_deltaker']])
```

```python
fields = ['person_id', 'ID_deltaker', 'TP_nr', 
          'Utgaar_sikkert', 'Utgaar_forelopig', 
          'Utgaar_lab', 'TYPE', 'GRAD', 'GRAD1', 
          'GRAD2', 'GRAD3', 'DUCTAL', 'LOBULAR', 
          'OTHER', 'EXCLUDED', 'Luminal', 
          'NonLuminal', 'HER2posgr', 'HER2neggr']
```

```python
new_df = pd.DataFrame()
for f in fields:
    new_df[f] = data.get(f)
```

```python
new_df
```

```python
np.all(np.isfinite(new_df) == (new_df == data.loc[:, fields]))
```

```python
new_df.to_stata('data/clinic_data.dta')
```
