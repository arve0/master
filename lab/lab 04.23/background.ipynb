```python
from skimage import io
```

```python
img = io.imread('background.tif')
```

```python
io.imsave('background.png', img)
```

```python
from IPython import display
```

```python
display.Image('background.png')
```

```python
from skimage import filters, morphology
```

```python
selem = morphology.square(50)
smooth = filters.median(img, selem)
```

```python
io.imsave('background-smooth.png', smooth)
display.Image('background-smooth.png')
```
